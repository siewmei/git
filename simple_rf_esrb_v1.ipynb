```python
# import necessary packages

import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import os
import pandas as pd
```

```python
# set the path and data filename

path = "C:\\Users\\siewm\\OneDrive\\Desktop\\DNN practice\\data\\"
datafile = 'Video_games_esrb_rating.csv'
testdata = 'test_esrb.csv'

os.chdir(path)

# read in data

d = pd.read_csv(path+datafile)
td = pd.read_csv(path+testdata)
```

```python
d
```

```python
td
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
# prepare train and test data

xtrain = d.drop(['title', 'esrb_rating'], axis=1)
ytrain = d['esrb_rating']
xtest = td.drop(['title', 'esrb_rating'], axis=1)
ytest = td['esrb_rating']
```

```python
# fit the train data into the rf model

rf.fit(xtrain, ytrain)
```

```python
# predict esrb_rating using test data

ypred = rf.predict(xtest)
```

```python
# assess accuracy of the rf classifier (eg. how often is the classifier correct?)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(ytest, ypred))
```

```python
# count how many is correct
accuracy = pd.DataFrame(ypred == ytest)
```
