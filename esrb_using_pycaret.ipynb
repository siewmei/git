1. Import packages needed

```python
import pandas as pd
from pycaret.classification import *
```

```python
# import the data into pd.DataFrame

import io
train = pd.read_csv(io.BytesIO(uploaded['Video_games_esrb_rating.csv']))
test = pd.read_csv(io.BytesIO(uploaded['test_esrb.csv']))

# Dataset is now stored in a Pandas Dataframe
```

4. Train and Evaluate Model

```python
# Prepare data, no need to split train and test because the setup function will perform the split.
# still, I will prepare two sets - train and test (for predicting later)

data = pd.concat([train, test], ignore_index=True)
data1 = data.drop('title', axis=1)
dtrain = train.drop('title', axis=1)
dtest = test.drop('title', axis=1)
```

```python
dtrain
```

```python
# Setting up
# setup function also performs the train test split (stratified for classification problems). 
# The default split ratio is 70:30 but you can change that using train_size parameter within setup. 
# Evaluation of a trained machine learning model and hyperparameter optimization in PyCaret is performed using k-fold cross validation on Train set only.
# must press 'ENTER' after confirming the variables

ex_models = setup(dtrain, target = 'esrb_rating')
```

```python
# this will return the best models excluding logistics regression
# it doesn't make sense to use logistics regression on multiclassification

bestmodels = compare_models(exclude=['lr'])

# return the top 3 best models
# best3 = compare_models(n_select = 3)

# return the best model based on AUC
# bestAUC = compare_models(sort = 'AUC')

# compare specific models
# models = compare_models(include = ['rf', 'xgboost', 'lightgbm', 'gbc'])
```

5. Predict using unseen data

```python
predict_model(bestmodels, data = dtest)
```
